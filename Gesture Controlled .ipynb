```python
# -------------------------------
# 2. INFERENCE WITH WEBCAM
# -------------------------------
import cv2

# Load trained model
model = tf.keras.models.load_model("gesture_sign_mnist.h5")

# Class mapping
class_names = ["B", "F", "G", "L", "R", "S"]
command_map = {
    "B": "BACKWARD",
    "F": "FORWARD",
    "G": "START",
    "L": "LEFT",
    "R": "RIGHT",
    "S": "STOP"
}

# Start webcam
cap = cv2.VideoCapture(1)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Define ROI (region of interest)
    roi = frame[100:300, 100:300]  # adjust as needed
    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    resized = cv2.resize(gray, (28, 28))
    reshaped = resized.reshape(1, 28, 28, 1) / 255.0

    # Predict
    preds = model.predict(reshaped)
    class_id = np.argmax(preds)
    gesture = class_names[class_id]
    command = command_map[gesture]

    # Display
    cv2.rectangle(frame, (100, 100), (300, 300), (0, 255, 0), 2)
    cv2.putText(frame, f"{gesture} -> {command}", (50, 50),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2)

    cv2.imshow("Gesture Recognition", frame)

    # Exit on 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
